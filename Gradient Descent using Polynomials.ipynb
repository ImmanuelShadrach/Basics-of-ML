```python
import numpy as np
import matplotlib.pyplot as plt

# Define the convex function and its derivative
def f(x):
    return x**2

def df(x):
    return 2*x

# Gradient descent function
def gradient_descent(f, df, learning_rate=0.01, tolerance=1e-5, max_iters=1000):
    # Initialize the starting point randomly
    x = np.random.uniform(-2, 2)
    #x=2.22

    # Lists to store the history of points and function values
    x_history = [x]
    f_history = [f(x)]

    # Gradient descent iterations
    for _ in range(max_iters):
        # Update the point using the gradient descent update rule
        x = x - learning_rate * df(x)

        # Append the updated point and function value to the history lists
        x_history.append(x)
        f_history.append(f(x))

        # Check for convergence
        if abs(f_history[-1] - f_history[-2]) < tolerance:
            break

    return x, x_history, f_history

# Run gradient descent
saddle_point, x_history, f_history = gradient_descent(f, df)

# Print the minimum point found
print("Minimum point found:", saddle_point)

# Plot the function and the gradient descent trajectory
x_values = np.linspace(-2.5, 2.5, 100)
plt.plot(x_values, f(x_values), label='Function')
plt.scatter(x_history, f_history, c='red', label='Gradient Descent Trajectory')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent to Minimize Convex Function')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its gradient
def f(x, y):
    return x**2 + y**2

def grad_f(x, y):
    return np.array([2*x, 2*y])

# Gradient descent function
def gradient_descent_2d(f, grad_f, learning_rate=0.1, tolerance=1e-5, max_iters=1000):
    # Initialize the starting point randomly
    x = np.random.uniform(-2, 2)
    y = np.random.uniform(-2, 2)

    # Lists to store the history of points and function values
    x_history = [x]
    y_history = [y]
    f_history = [f(x, y)]

    # Gradient descent iterations
    for _ in range(max_iters):
        # Compute the gradient at the current point
        grad = grad_f(x, y)

        # Update the point using the gradient descent update rule
        x = x - learning_rate * grad[0]
        y = y - learning_rate * grad[1]

        # Append the updated point and function value to the history lists
        x_history.append(x)
        y_history.append(y)
        f_history.append(f(x, y))

        # Check for convergence
        if abs(f_history[-1] - f_history[-2]) < tolerance:
            break

    return x_history, y_history, f_history

# Run gradient descent
x_history, y_history, f_history = gradient_descent_2d(f, grad_f)

# Plot the contour plot of the function
x = np.linspace(-2.5, 2.5, 100)
y = np.linspace(-2.5, 2.5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contourf(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar(label='Function Value')

# Plot the convergence trajectory
plt.plot(x_history, y_history, marker='o', color='red', label='Convergence')

# Annotate the starting and ending points
plt.annotate('Start', xy=(x_history[0], y_history[0]), xytext=(-2, -2),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('End', xy=(x_history[-1], y_history[-1]), xytext=(2, 2),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Convergence for 2D Minimization')
plt.legend()
plt.grid(True)
plt.show()
```
